# Obtenció dades estació Sabadell Parc Agrari

## Temps estimat ~ 2 hores 25 min

```python
import requests
import pandas as pd
from tqdm import tqdm
from io import StringIO
import time

import warnings
warnings.filterwarnings("ignore")

# -----------------------------
# Configuration
# -----------------------------
station_code = "XF"
start_date = "2008-10-24"
end_date = "2026-04-01"
output_file = "Sbd_nord_data.csv"
failed_file = "Sbd_nord_failed_dates.csv"

headers = {
    "User-Agent": "Mozilla/5.0"
}

# -----------------------------
# Date range
# -----------------------------
dates = pd.date_range(start=start_date, end=end_date, freq="D")

# -----------------------------
# Storage
# -----------------------------
all_days = []
failed_dates = []

# -----------------------------
# Session
# -----------------------------
session = requests.Session()
session.headers.update(headers)

# -----------------------------
# Main loop
# -----------------------------
for day in tqdm(dates, desc="Downloading Meteocat data"):
    day_str = day.strftime("%Y-%m-%d")
    url = f"https://www.meteo.cat/observacions/xema/dades?codi={station_code}&dia={day_str}T00:00Z"

    try:
        response = session.get(url, timeout=20)
        response.raise_for_status()

        tables = pd.read_html(StringIO(response.text))

        if len(tables) < 3:
            failed_dates.append((day_str, "Less than 3 tables found"))
            continue

        # Force a fully independent dataframe
        df = pd.DataFrame(tables[2]).copy(deep=True)

        df = df.rename(columns={
            "PeríodeTU": "period_utc",
            "TM°C": "temp_mean (°C)",
            "TX°C": "temp_max (°C)",
            "TN°C": "temp_min (°C)",
            "HRM%": "humidity (%)",
            "PPTmm": "precip (mm)",
            "VVM (10 m)km/h": "wind_mean_10m (km/h)",
            "DVM (10 m)graus": "wind_dir_10m (degrees)",
            "VVX (10 m)km/h": "wind_max_10m (km/h)",
            "PMhPa": "pressure (hPa)",
            "RSW/m2": "radiation (W/m²)"
        }).copy(deep=True)

        # Build new columns using assign to avoid warning
        df = df.assign(
            time_utc=df["period_utc"].astype(str).str.split(" - ").str[0]
        )

        df = df.assign(
            datetime_utc=pd.to_datetime(
                day_str + " " + df["time_utc"],
                errors="coerce"
            )
        )

        df = df.dropna(subset=["datetime_utc"]).copy(deep=True)

        df = df.drop(columns=["period_utc", "time_utc"])
        df = df.set_index("datetime_utc")

        numeric_cols = df.columns
        df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors="coerce")

        df["station_code"] = station_code

        all_days.append(df)

        time.sleep(0.2)

    except Exception as e:
        failed_dates.append((day_str, str(e)))


path = "C:\\Users\\David\\Documents\\VO project\\VO-evolution\\Raw Data"
# -----------------------------
# Combine and save
# -----------------------------
if all_days:
    final_df = pd.concat(all_days).sort_index().reset_index()
    final_df.to_csv(path + "\\" + output_file, index=False, encoding="utf-8")
    print(f"Saved {len(final_df)} rows to '{output_file}'")
else:
    print("No data was collected.")

# -----------------------------
# Save failed dates
# -----------------------------
if failed_dates:
    failed_df = pd.DataFrame(failed_dates, columns=["date", "error"])
    failed_df.to_csv(path + "\\" + failed_file, index=False, encoding="utf-8")
    print(f"Saved {len(failed_df)} failed dates to '{failed_file}'")
else:
    print("No failed dates.")
```

# Obtenció dades SBD centre desde 2008 fins a 2018

```python
import requests
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
from datetime import datetime

import warnings
warnings.filterwarnings("ignore", category=FutureWarning)

MONTH_NAMES = {
    1: "January",
    2: "February",
    3: "March",
    4: "April",
    5: "May",
    6: "June",
    7: "July",
    8: "August",
    9: "September",
    10: "October",
    11: "November",
    12: "December",
}

HEADERS = {
    0: "day",
    1: "max",
    2: "min",
    3: "avg",
    4: "dep_from_normal",
    5: "dew_point",
    6: "wet_bulb",
    7: "heating",
    8: "cooling",
    9: "sig_cond",
    10: "sun_hrs",
    11: "snow_fall",
    12: "rain",
    13: "pressure",
    14: "wind_speed",
    15: "wind_dir",
    16: "wind_max",
    17: "wind_max_dir",
    # 18 és l'última columna "day" repetida a la dreta, no cal
}

NUMERIC_COLS = [
    "day", "max", "min", "avg", "dep_from_normal", "dew_point", "wet_bulb",
    "heating", "cooling", "sun_hrs", "snow_fall", "rain", "pressure",
    "wind_speed", "wind_dir", "wind_max", "wind_max_dir"
]


def build_url(year: int, month: int) -> str:
    month_name = MONTH_NAMES[month]
    return f"https://www.meteosabadell.cat/www/NOAA/{year}/{month_name}{year}/climatedataout{month}{year}.html"


def extract_column(td):
    vals = []
    for s in td.stripped_strings:
        s = s.replace("\xa0", " ").strip()
        if s:
            vals.append(s)
    return vals


def normalize_length(values, target_len):
    values = values[:target_len]
    if len(values) < target_len:
        values = values + [np.nan] * (target_len - len(values))
    return values


def find_main_data_row(rows):
    """
    Busca la fila principal de dades:
    la que té moltes columnes i on la primera conté els dies.
    """
    for row in rows:
        cols = row.find_all("td")
        if len(cols) >= 19:
            first_vals = extract_column(cols[0])
            digit_vals = [x for x in first_vals if x.isdigit()]
            if "1" in digit_vals and len(digit_vals) >= 28:
                return row
    return None


def clean_dataframe(df):
    df = df.replace({
        "-----": np.nan,
        "---": np.nan,
        "-": np.nan,
        "": np.nan
    })

    for col in ["wind_dir", "wind_max_dir"]:
        if col in df.columns:
            df[col] = df[col].astype(str).str.replace("~", "", regex=False)
            df[col] = df[col].replace("nan", np.nan)

    for col in NUMERIC_COLS:
        if col in df.columns:
            df[col] = pd.to_numeric(df[col], errors="coerce")

    if "sig_cond" in df.columns:
        df["sig_cond"] = df["sig_cond"].replace("nan", np.nan)

    return df


def parse_month(url: str, year: int, month: int, timeout: int = 30) -> pd.DataFrame:
    response = requests.get(url, timeout=timeout)
    response.raise_for_status()
    response.encoding = "ISO-8859-1"

    soup = BeautifulSoup(response.text, "html.parser")
    table = soup.find("table")
    if table is None:
        raise ValueError("No s'ha trobat cap taula a la pàgina.")

    rows = table.find_all("tr")
    data_row = find_main_data_row(rows)
    if data_row is None:
        raise ValueError("No s'ha trobat la fila principal de dades.")

    cols = data_row.find_all("td")
    if len(cols) < 18:
        raise ValueError(f"Nombre de columnes insuficient: {len(cols)}")

    # dies
    day_vals = extract_column(cols[0])
    day_vals = [x for x in day_vals if x.isdigit()]
    n_days = len(day_vals)

    # extreure totes les columnes útils
    parsed = {"day": day_vals}
    for idx, name in HEADERS.items():
        if idx == 0:
            continue
        parsed[name] = normalize_length(extract_column(cols[idx]), n_days)

    df = pd.DataFrame(parsed)
    df = clean_dataframe(df)

    df["year"] = year
    df["month"] = month
    df["date"] = pd.to_datetime(
        dict(year=df["year"], month=df["month"], day=df["day"]),
        errors="coerce"
    )

    # Reordenar
    ordered_cols = [
        "date", "year", "month", "day",
        "max", "min", "avg", "dep_from_normal", "dew_point", "wet_bulb",
        "heating", "cooling", "sig_cond", "sun_hrs", "snow_fall", "rain",
        "pressure", "wind_speed", "wind_dir", "wind_max", "wind_max_dir"
    ]
    df = df[ordered_cols]

    return df


def month_range(start_year: int, start_month: int, end_year: int, end_month: int):
    y, m = start_year, start_month
    while (y < end_year) or (y == end_year and m <= end_month):
        yield y, m
        m += 1
        if m == 13:
            m = 1
            y += 1


all_dfs = []
errors = []

for year, month in month_range(2008, 5, 2018, 1):
    url = build_url(year, month)
    print(f"Processant {year}-{month:02d} ...")

    try:
        df_month = parse_month(url, year, month)
        all_dfs.append(df_month)
        print(f"  OK -> {len(df_month)} files")
    except Exception as e:
        print(f"  ERROR -> {e}")
        errors.append({
            "year": year,
            "month": month,
            "url": url,
            "error": str(e)
        })

# Concatenar tots els mesos
if all_dfs:
    df_all = pd.concat(all_dfs, ignore_index=True)
    df_all = df_all.sort_values("date").reset_index(drop=True)
else:
    df_all = pd.DataFrame()

# Guardar resultats
df_all.to_csv("meteosabadell_2008-05_to_2018-01.csv", index=False)

errors_df = pd.DataFrame(errors)
errors_df.to_csv("meteosabadell_2008-05_to_2018-01_errors.csv", index=False)

print("\n==============================")
print(f"Files totals: {len(df_all)}")
print(f"Mesos correctes: {len(all_dfs)}")
print(f"Mesos amb error: {len(errors)}")
print("==============================")
```

# Obtenció SBD Centre 2018 - 2026

```python
import requests
import pandas as pd
import re
from bs4 import BeautifulSoup
from datetime import datetime
from io import StringIO

# =========================
# CONFIG
# =========================

BASE_URL = "http://www.meteosabadell.cat/www/NOAA"
START_YEAR = 2018
START_MONTH = 1
END_YEAR = 2026
END_MONTH = 1

OUTPUT_DAILY_CSV = "meteosabadell_2018-01_to_2026-01.csv"
OUTPUT_ERRORS_CSV = "meteosabadell_parse_errors.csv"

TIMEOUT = 20

session = requests.Session()
session.headers.update({
    "User-Agent": "Mozilla/5.0 (compatible; MeteoSabadellParser/1.0)"
})

# =========================
# MONTH MAPS
# =========================

CAT_MONTHS = {
    1: "Gener",
    2: "Febrer",
    3: "Mar",
    4: "Abril",
    5: "Maig",
    6: "Juny",
    7: "Juliol",
    8: "Agost",
    9: "Setembre",
    10: "Octubre",
    11: "Novembre",
    12: "Desembre",
}

ENG_MONTHS_FULL = {
    1: "January",
    2: "February",
    3: "March",
    4: "April",
    5: "May",
    6: "June",
    7: "July",
    8: "August",
    9: "September",
    10: "October",
    11: "November",
    12: "December",
}

ENG_MONTHS_NOAA_2018 = {
    1: "January2018",  # cas especial vist a la captura
}

# =========================
# HELPERS
# =========================

def month_iter(start_year, start_month, end_year, end_month):
    y, m = start_year, start_month
    while (y < end_year) or (y == end_year and m <= end_month):
        yield y, m
        m += 1
        if m == 13:
            m = 1
            y += 1

def safe_get(url):
    try:
        r = session.get(url, timeout=TIMEOUT)
        if r.status_code == 200:
            return r
        return None
    except requests.RequestException:
        return None

def to_float(value):
    if value is None:
        return None
    value = value.strip().replace(",", ".")
    value = value.replace("~", "")
    value = value.replace("O", "0")
    value = value.replace("o", "0")
    value = value.replace("—", "").replace("--", "").replace("-----", "")
    if value in {"", "----", "-----", "------"}:
        return None
    try:
        return float(value)
    except ValueError:
        return None

def to_int(value):
    if value is None:
        return None
    value = value.strip()
    value = value.replace("O", "0").replace("o", "0")
    if not value.isdigit():
        return None
    return int(value)

def normalize_time(value):
    if value is None:
        return None
    value = value.strip().replace(";", ":")
    m = re.match(r"^(\d{1,2}):(\d{2})$", value)
    if not m:
        return None
    hh = int(m.group(1))
    mm = int(m.group(2))
    if 0 <= hh <= 23 and 0 <= mm <= 59:
        return f"{hh:02d}:{mm:02d}"
    return None

def normalize_wind_dir(value):
    if value is None:
        return None
    value = value.strip().upper().replace("5", "S")
    valid = {
        "N", "NNE", "NE", "ENE", "E", "ESE", "SE", "SSE",
        "S", "SSW", "SW", "WSW", "W", "WNW", "NW", "NNW"
    }
    return value if value in valid else value

def make_iso_date(year, month, day):
    try:
        return datetime(year, month, day).strftime("%Y-%m-%d")
    except ValueError:
        return None

# =========================
# URL BUILDERS
# =========================

def candidate_urls(year, month):
    urls = []

    # Format nou TXT (vist per 2018+)
    cat_month = CAT_MONTHS[month]
    urls.append(f"{BASE_URL}/{year}/{cat_month}_{year}_NOAA.TXT")

    # Alguns fitxers poden tenir variants de nom
    if month == 3:
        urls.append(f"{BASE_URL}/{year}/Mar_{year}_NOAA.TXT")

    # Format HTML antic
    eng_month = ENG_MONTHS_FULL[month]
    urls.append(f"{BASE_URL}/{year}/{eng_month}{year}/climatedataout{month}{year}.html")

    # Cas especial gener 2018 vist a la captura
    if year == 2018 and month == 1:
        urls.append(f"{BASE_URL}/{year}/January2018/climatedataout12018.html")

    return urls

def find_existing_url(year, month):
    for url in candidate_urls(year, month):
        r = safe_get(url)
        if r is not None:
            return url, r
    return None, None

# =========================
# NOAA TXT PARSER
# =========================

TXT_ROW_PATTERN = re.compile(
    r'^\s*(\d{1,2})\s+'              # day
    r'(-?\d+\.\d)\s+'                # mean temp
    r'(-?\d+\.\d)\s+'                # high
    r'(\d{1,2}:\d{2})\s+'            # high time
    r'(-?\d+\.\d)\s+'                # low
    r'(\d{1,2}:\d{2})\s+'            # low time
    r'(-?\d+\.\d)\s+'                # heat deg days
    r'(-?\d+\.\d)\s+'                # cool deg days
    r'(-?\d+\.\d)\s+'                # rain
    r'(-?\d+\.\d)\s+'                # avg wind speed
    r'(-?\d+\.\d)\s+'                # high wind
    r'(\d{1,2}:\d{2})\s+'            # high wind time
    r'([A-Z]{1,3})\s*$'              # dom dir
)

def parse_noaa_txt(text, year, month, source_url):
    rows = []
    for line in text.splitlines():
        m = TXT_ROW_PATTERN.match(line)
        if not m:
            continue

        day = int(m.group(1))
        row = {
            "date": make_iso_date(year, month, day),
            "year": year,
            "month": month,
            "day": day,
            "source_format": "txt_noaa",
            "source_url": source_url,

            "mean_temp_c": float(m.group(2)),
            "max_temp_c": float(m.group(3)),
            "max_temp_time": normalize_time(m.group(4)),
            "min_temp_c": float(m.group(5)),
            "min_temp_time": normalize_time(m.group(6)),
            "heat_deg_days": float(m.group(7)),
            "cool_deg_days": float(m.group(8)),
            "rain_mm": float(m.group(9)),
            "avg_wind_kmh": float(m.group(10)),
            "max_wind_kmh": float(m.group(11)),
            "max_wind_time": normalize_time(m.group(12)),
            "dom_wind_dir": normalize_wind_dir(m.group(13)),
        }
        rows.append(row)
    return rows

# =========================
# HTML PARSER
# =========================

def clean_cell_text(td):
    text = td.get_text(" ", strip=True)
    text = text.replace("\xa0", " ").strip()
    return text

def parse_html_table_rows(soup):
    """
    Busca files de taula candidates que comencin per un dia 1..31.
    """
    candidate_rows = []
    for tr in soup.find_all("tr"):
        cells = tr.find_all(["td", "th"])
        texts = [clean_cell_text(c) for c in cells]
        if not texts:
            continue

        first = texts[0].strip()
        if re.fullmatch(r"\d{1,2}", first):
            day = int(first)
            if 1 <= day <= 31:
                candidate_rows.append(texts)

    return candidate_rows

def parse_old_html(html, year, month, source_url):
    soup = BeautifulSoup(html, "html.parser")
    raw_rows = parse_html_table_rows(soup)

    parsed = []

    for cells in raw_rows:
        # Aquest format antic acostuma a tenir moltes columnes.
        # Ens quedem amb les més útils i relativament estables.
        #
        # Estructura observada aproximada:
        # 0 day
        # 1 max temp
        # 2 min temp
        # 3 avg temp
        # 4 dep from normal
        # 5 avg dew pt
        # 6 avg wet bulb
        # 7 heating
        # 8 cooling
        # 9 sig cond
        # 10 sun hrs
        # 11 snow fall
        # 12 water equiv / rain
        # 13 avg sea level pressure
        # 14 avg spd
        # 15 avg dir
        # 16 max spd
        # 17 max spd dir
        # 18 day (repetit al final)

        if len(cells) < 17:
            continue

        day = to_int(cells[0])
        if day is None:
            continue

        row = {
            "date": make_iso_date(year, month, day),
            "year": year,
            "month": month,
            "day": day,
            "source_format": "html_old",
            "source_url": source_url,

            "mean_temp_c": to_float(cells[3]) if len(cells) > 3 else None,
            "max_temp_c": to_float(cells[1]) if len(cells) > 1 else None,
            "max_temp_time": None,
            "min_temp_c": to_float(cells[2]) if len(cells) > 2 else None,
            "min_temp_time": None,
            "heat_deg_days": to_float(cells[7]) if len(cells) > 7 else None,
            "cool_deg_days": to_float(cells[8]) if len(cells) > 8 else None,
            "rain_mm": to_float(cells[12]) if len(cells) > 12 else None,
            "avg_wind_kmh": to_float(cells[14]) if len(cells) > 14 else None,
            "max_wind_kmh": to_float(cells[16]) if len(cells) > 16 else None,
            "max_wind_time": None,
            "dom_wind_dir": cells[15].strip() if len(cells) > 15 else None,

            # columnes extra del format antic
            "dep_from_normal_c": to_float(cells[4]) if len(cells) > 4 else None,
            "avg_dew_point_c": to_float(cells[5]) if len(cells) > 5 else None,
            "avg_wet_bulb_c": to_float(cells[6]) if len(cells) > 6 else None,
            "sig_condition": cells[9].strip() if len(cells) > 9 else None,
            "sun_hours": to_float(cells[10]) if len(cells) > 10 else None,
            "snow_fall_cm": to_float(cells[11]) if len(cells) > 11 else None,
            "avg_sea_level_pressure_hpa": to_float(cells[13]) if len(cells) > 13 else None,
            "max_wind_dir_deg": cells[17].strip() if len(cells) > 17 else None,
        }

        parsed.append(row)

    return parsed

# =========================
# MAIN
# =========================

def parse_month(year, month):
    url, response = find_existing_url(year, month)

    if url is None or response is None:
        return [], {
            "year": year,
            "month": month,
            "url": None,
            "status": "not_found",
            "message": "No s'ha trobat cap URL vàlida"
        }

    try:
        if url.lower().endswith(".txt"):
            rows = parse_noaa_txt(response.text, year, month, url)
        elif url.lower().endswith(".html"):
            rows = parse_old_html(response.text, year, month, url)
        else:
            return [], {
                "year": year,
                "month": month,
                "url": url,
                "status": "unsupported_format",
                "message": "Format no suportat"
            }

        if not rows:
            return [], {
                "year": year,
                "month": month,
                "url": url,
                "status": "parsed_zero_rows",
                "message": "Fitxer trobat però no s'han extret files"
            }

        return rows, None

    except Exception as e:
        return [], {
            "year": year,
            "month": month,
            "url": url,
            "status": "parse_error",
            "message": str(e)
        }

def main():
    all_rows = []
    errors = []

    for year, month in month_iter(START_YEAR, START_MONTH, END_YEAR, END_MONTH):
        print(f"Processant {year}-{month:02d} ...")
        rows, err = parse_month(year, month)

        if rows:
            all_rows.extend(rows)
            print(f"  OK -> {len(rows)} files")
        if err:
            errors.append(err)
            print(f"  ERROR -> {err['status']}")

    df = pd.DataFrame(all_rows)

    if not df.empty:
        # Reordena columnes
        preferred_cols = [
            "date", "year", "month", "day",
            "source_format", "source_url",
            "mean_temp_c", "max_temp_c", "max_temp_time",
            "min_temp_c", "min_temp_time",
            "heat_deg_days", "cool_deg_days",
            "rain_mm",
            "avg_wind_kmh", "max_wind_kmh", "max_wind_time", "dom_wind_dir",
            "dep_from_normal_c", "avg_dew_point_c", "avg_wet_bulb_c",
            "sig_condition", "sun_hours", "snow_fall_cm",
            "avg_sea_level_pressure_hpa", "max_wind_dir_deg"
        ]
        existing_cols = [c for c in preferred_cols if c in df.columns]
        remaining_cols = [c for c in df.columns if c not in existing_cols]
        df = df[existing_cols + remaining_cols]

        df = df.sort_values(["date", "day"], na_position="last").reset_index(drop=True)
        df.to_csv(OUTPUT_DAILY_CSV, index=False, encoding="utf-8")
        print(f"\nCSV diari desat a: {OUTPUT_DAILY_CSV}")
        print(f"Total files: {len(df)}")

    err_df = pd.DataFrame(errors)
    err_df.to_csv(OUTPUT_ERRORS_CSV, index=False, encoding="utf-8")
    print(f"CSV d'errors desat a: {OUTPUT_ERRORS_CSV}")
    print(f"Total errors/incidències: {len(err_df)}")

if __name__ == "__main__":
    main()
```
