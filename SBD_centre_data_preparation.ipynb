# Codi Per Concatenar data sbd centre

```python
import pandas as pd

df_2008 = pd.read_csv('meteosabadell_2008-05_to_2018-01.csv')
df_2008.head()
```

```python
df_2018 = pd.read_csv('meteosabadell_2018-01_to_2026-01.csv')
df_2018.head()
```

```python
df_2008.rename(columns = {'max':'T_max','min':'T_min','avg':'T_avg', 'rain':'rain_mm','wind_speed':'avg_wind_kmh','wind_max':'max_wind_kmh'}, inplace = True)
df_2018.rename(columns = {'max_temp_c':'T_max','min_temp_c':'T_min','mean_temp_c':'T_avg','dom_wind_dir':'wind_dir'}, inplace = True)

df_2008.drop(columns = ['dew_point','wet_bulb','cooling','sun_hrs','snow_fall','pressure','sig_cond'],axis = 1, inplace = True)
df_2018.drop(columns = ['source_format','source_url','max_temp_time','min_temp_time','heat_deg_days','cool_deg_days','max_wind_time'],axis = 1, inplace = True)
```

```python
import pandas as pd
import numpy as np

dir_to_deg = {
    'N': 0.0,
    'NNE': 22.5,
    'NE': 45.0,
    'ENE': 67.5,
    'E': 90.0,
    'ESE': 112.5,
    'SE': 135.0,
    'SSE': 157.5,
    'S': 180.0,
    'SSW': 202.5,
    'SW': 225.0,
    'WSW': 247.5,
    'W': 270.0,
    'WNW': 292.5,
    'NW': 315.0,
    'NNW': 337.5
}

def convert_wind_dir(x):
    if pd.isna(x):
        return np.nan
    
    # Si ja és numèric, el deixem com float
    if isinstance(x, (int, float, np.integer, np.floating)):
        return float(x)
    
    # Si és string, netegem espais
    x = str(x).strip().upper()
    
    # Si és una direcció cardinal
    if x in dir_to_deg:
        return dir_to_deg[x]
    
    # Si és un número guardat com string
    try:
        return float(x)
    except ValueError:
        return np.nan

df_sbd_centre['wind_dir_deg'] = df_sbd_centre['wind_dir'].apply(convert_wind_dir)
```

```python
df_sbd_centre[df_sbd_centre['wind_dir_deg'].isna()]['wind_dir'].unique()
```

```python
df_sbd_centre['wind_dir'] = df_sbd_centre['wind_dir'].apply(convert_wind_dir)
```

```python
df_sbd_centre.drop(columns = ['wind_dir_deg'], axis = 1, inplace = True)
```
