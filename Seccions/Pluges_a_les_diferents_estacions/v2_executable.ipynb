# Pluges a les diferents estacions - v2 ( executable )

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
import warnings
warnings.filterwarnings("ignore")
# Optional: display plots inline if running in Jupyter
```

```python
PRIMARY_BASE = "https://raw.githubusercontent.com/DavidDDRC99/VO-evolution/refs/heads/main/Data/Cleaned%20Data/"
FALLBACK_BASE = "https://raw.githubusercontent.com/DavidDDRC99/VO-evolution/main/Data/Cleaned%20Data/"
NAMES = [
  "Sbd_nord_daily.csv",
  "Sbd_Centre_daily.csv",
  "Vacarisses_daily.csv"
]
```

```python
def load_csv_with_fallback(name):
    urls = [f'{PRIMARY_BASE}{name}.csv', FALLBACK_BASE + f'{name}.csv']
    for url in urls:
        try:
            df = pd.read_csv(url)
            print(f'Loaded {name} from {url}')
            return df
        except Exception as e:
            print(f'Failed {name} from {url}: {e}')
            continue
    return pd.DataFrame()

def standardize_columns(df, station_label):
    target_cols = ['date','year','month','day','T_max','T_min','T_avg','rain_mm','avg_wind_kmh','wind_dir','max_wind_kmh','wind_max_dir']
    if 'date' in df.columns and not pd.api.types.is_datetime64_any_dtype(df['date']):
        df['date'] = pd.to_datetime(df['date'], errors='coerce')
    for c in target_cols:
        if c not in df.columns:
            df[c] = np.nan
    df['station'] = station_label
    return df
```

```python
df_parc = load_csv_with_fallback('Sbd_nord_daily')
df_centre = load_csv_with_fallback('Sbd_Centre_daily')
df_vacarisses = load_csv_with_fallback('Vacarisses_daily')
df_parc_s = standardize_columns(df_parc, 'Sbd_nord_daily')
df_centre_s = standardize_columns(df_centre, 'Sbd_Centre_daily')
df_vacarisses_s = standardize_columns(df_vacarisses, 'Vacarisses_daily')
df_all = pd.concat([df_parc_s, df_centre_s, df_vacarisses_s], ignore_index=True)
df_all.to_csv('Seccions/Pluges_a_les_diferents_estacions/v2_Pluges_consolidated.csv', index=False)
print('Consolidated CSV written')
plt.figure(figsize=(12,6))
for st in df_all['station'].dropna().unique():
    sub = df_all[df_all['station']==st]
    if 'date' in sub.columns and 'rain_mm' in sub.columns:
        plt.plot(pd.to_datetime(sub['date']), sub['rain_mm'], label=st)
plt.xlabel('Date')
plt.ylabel('Rain (mm)')
plt.legend()
plt.tight_layout()
plt.show()
```
