```python
import pandas as pd
import numpy as np

dir_to_deg = {
    'N': 0.0,
    'NNE': 22.5,
    'NE': 45.0,
    'ENE': 67.5,
    'E': 90.0,
    'ESE': 112.5,
    'SE': 135.0,
    'SSE': 157.5,
    'S': 180.0,
    'SSW': 202.5,
    'SW': 225.0,
    'WSW': 247.5,
    'W': 270.0,
    'WNW': 292.5,
    'NW': 315.0,
    'NNW': 337.5
}

def convert_wind_dir(x):
    if pd.isna(x):
        return np.nan
    
    # Si ja és numèric, el deixem com float
    if isinstance(x, (int, float, np.integer, np.floating)):
        return float(x)
    
    # Si és string, netegem espais
    x = str(x).strip().upper()
    
    # Si és una direcció cardinal
    if x in dir_to_deg:
        return dir_to_deg[x]
    
    # Si és un número guardat com string
    try:
        return float(x)
    except ValueError:
        return np.nan

df_sbd_centre['wind_dir_deg'] = df_sbd_centre['wind_dir'].apply(convert_wind_dir)
```

```python
df_sbd_centre[df_sbd_centre['wind_dir_deg'].isna()]['wind_dir'].unique()
```

```python
df_sbd_centre['wind_dir'] = df_sbd_centre['wind_dir'].apply(convert_wind_dir)
```

```python
df_sbd_centre.drop(columns = ['wind_dir_deg'], axis = 1, inplace = True)
```

# Transformació de les dades de Vacarisses: Horari ---> Diari

```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/DavidDDRC99/VO-evolution/refs/heads/main/Data/Vacarisses_data.csv')
df.head()
```

```python
df.drop(columns = ['station_code'], inplace = True)
df.head()
```

```python
import pandas as pd
import numpy as np

# 1. Assegurar datetime
df["datetime_utc"] = pd.to_datetime(df["datetime_utc"], errors="coerce")

# 2. Funció per direcció del vent (important!)
def mean_wind_dir(deg):
    deg = deg.dropna()
    if len(deg) == 0:
        return np.nan
    rad = np.deg2rad(deg)
    sin_mean = np.mean(np.sin(rad))
    cos_mean = np.mean(np.cos(rad))
    return np.rad2deg(np.arctan2(sin_mean, cos_mean)) % 360

# 3. Agregació diària
daily_df = (
    df.groupby(df["datetime_utc"].dt.floor("D"))
      .agg(
          T_avg=("temp_mean (°C)", "mean"),
          T_max=("temp_max (°C)", "max"),
          T_min=("temp_min (°C)", "min"),
          humidity_avg=("humidity (%)", "mean"),
          precip_total=("precip (mm)", "sum"),
          wind_avg=("wind_mean_10m (km/h)", "mean"),
          wind_dir=("wind_dir_10m (degrees)", mean_wind_dir),
          wind_max=("wind_max_10m (km/h)", "max"),
          pressure_avg=("pressure (hPa)", "mean"),
          radiation_avg=("radiation (W/m²)", "mean"),
      )
      .reset_index()
)

# 4. Rename columna data
daily_df = daily_df.rename(columns={"datetime_utc": "date"})
```

```python
daily_df['radiation_avg'].isna().sum()
```

```python
daily_df.rename(columns = {'precip_total':'rain_mm', 'wind_avg':'avg_wind_kmh','wind_max':'max_wind_kmh'}, inplace = True)
```

```python
nan_percent = daily_df.isna().mean() * 100

print(nan_percent)
```

```python
daily_df.drop(columns=['radiation_avg','max_wind_kmh','wind_dir','avg_wind_kmh'], inplace=True)
```
